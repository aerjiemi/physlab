```python
import pandas as pd

df = pd.read_excel("lab11.xlsx")
df.columns
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(10, 6))

t_data = df['T2']
u_data = df['U3']
p_data = df['U3'] * df['I1']
p_avg = p_data.mean()

# 
ax1.plot(t_data, u_data, color='red', label='U1(t)', linewidth=1.5, alpha=0.8)
ax1.set_xlabel(' T, ')
ax1.set_ylabel(' U1, ', color='red')
ax1.tick_params(axis='y', labelcolor='red')
ax1.grid(True, which='both', linestyle='--', alpha=0.5)

# 
ax2 = ax1.twinx()
ax2.plot(t_data, p_data, color='green', label='P(t)', linewidth=2)
ax2.axhline(y=p_avg, color='darkorange', linestyle=':', label=f'P_ = {p_avg:.3f} ')
ax2.set_ylabel(' P, ', color='green')
ax2.tick_params(axis='y', labelcolor='green')


p_max = p_data.max()
u_max = u_data.max()
info_text = (f"U_max = {u_max:.2f} \n"
             f"P_max = {p_max:.3f} \n"
             f"P_avg = {p_avg:.5f} ")

plt.text(0.05, 0.95, info_text, transform=ax1.transAxes, fontsize=10,
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.7))


lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

plt.title('   U(t)   P(t)')
fig.tight_layout()

plt.savefig('U_P_plot.png')
plt.show()
```
